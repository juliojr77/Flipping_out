```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random
import itertools
```

```python
def flip_coin():
    return random.randint(0, 1)
        

def diff_flip_coin_n_times(times):  
    h = 0
    t = 0
    record_intrv_result = []
    intrv_rintrv_ratios = []
    
    for _ in range(int(times)):
        if flip_coin() == 0:
            h += 1
        else:
            t += 1     
        record_intrv_result.append(h - t) 

        if h == 0:
            intrv_rintrv_ratios.append(0)
            continue       
        elif t == 0:
            intrv_rintrv_ratios.append(0)
            continue
        else:   
            if h == min(h,t): 
                intrv_rintrv_ratios.append(round(1/(h / t),2))
            else:                   
                intrv_rintrv_ratios.append(round(1/(h / t),2))

    return record_intrv_result, intrv_rintrv_ratios
```

```python

def plot_flip_coin_n_times(n_times=2**8):
    interval = 0
    results = ()
    ratios = ()
    plt.figure(figsize=(20, 7))
    interval_2 = 0
    
    while 2**interval < n_times: 
        plt.plot(diff_flip_coin_n_times(2**interval)[0], label='Interval: '+ str(interval))
        plt.title("Diference between Heads and Tails\n (h - t)\n", fontsize=35, color='blue')
        plt.xlabel("x - Axis")
        plt.ylabel("y - Axis")
        plt.legend()
        interval += 1
    print('Total amount of flips: ' + str(n_times))
    print('Ammount of intervals: ' + str(interval))
                   
    plt.figure(figsize=(20, 7)) 

    while 2**interval_2 < n_times:     
        plt.plot(diff_flip_coin_n_times(2**interval)[1], label='Interval: '+ str(interval_2))
        plt.title("Ratio of Heads and Tails\n ", fontsize=35, color=(1.0,0.2,0.3))
        plt.xlabel("x - Axis", fontsize=20)
        plt.ylabel("y - Axis", fontsize=20)
        plt.legend()
        interval_2 += 1   

 
```

```python
plot_flip_coin_n_times()
```
